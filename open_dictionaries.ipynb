```python
import json
```

```python
city = "Los Angeles"
```

```python
with open('areas.txt', 'r') as f:
    areas = json.loads(f.read())
print("Total number of cities =", len(areas.keys()))
print("Area {} =".format(city), areas[city])
```

```python
import json

with open('tracts.txt', 'r') as f:
    tracts = json.loads(f.read())
print("Total number of tracts {} =".format(city), len(tracts[city]))
print("Tracts {} =".format(city), tracts[city])
```
