```python
import matplotlib.pyplot as plt
from matplotlib import patches
import numpy as np
import pandas as pd
```

```python
results_original_original = pd.read_csv('UC_results_original_features_original_data.csv', skiprows=1)
```

```python
print(results_original_original)
```

```python
SVM = results_original_original.iloc[0]
print(SVM)
```

```python
def create_handle(label_name, color, m):
    handle = plt.scatter([0], [0], label=label_name, c=color, marker=m)
    plt.close()
    return handle
```

```python
print(results_original_original.iloc[23])
```

```python
cities = ['Los Angeles', 'Austin', 'New Orleans', 'New York', 'Oakland', 'San Diego', 'San Francisco', 'Seattle']
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:cyan']
markers = ['x', '1', '2', '|', '+']
methods = ['None', 'Undersampling', 'Oversampling', 'Ensemble 3', 'Ensemble 5']
models = ['SVM', 'LR', 'RF', 'NB', 'EM', 'NN']
```

```python
handles = []
labels = []
for i in range(len(cities)):
    handles.append(create_handle(cities[i], colors[i], '.'))
    labels.append(cities[i])
r = patches.Rectangle((0,0), 1, 1, fill=False, edgecolor='none',
                                 visible=False)
handles.append(r)
labels.append('')
r = patches.Rectangle((0,0), 1, 1, fill=False, edgecolor='none',
                                 visible=False)
handles.append(r)
labels.append('')
for i in range(len(markers)):
    handles.append(create_handle(methods[i], 'black', markers[i]))
    labels.append(methods[i])
```

```python
fig = plt.figure()
ax = plt.subplot(111)
for i in range(len(markers)):
    print(4*i, 1+(4*i), 2+(4*i), 3+(4*i))
    ax.scatter([i]*len(cities), list(results_original_original.iloc[4*i])[1:], c=colors, marker=markers[i])
    ax.scatter([i+10]*len(cities), list(results_original_original.iloc[1+(4*i)])[1:], c=colors, marker=markers[i])
    ax.scatter([i+20]*len(cities), list(results_original_original.iloc[2+(4*i)])[1:], c=colors, marker=markers[i])
    ax.scatter([i+30]*len(cities), list(results_original_original.iloc[3+(4*i)])[1:], c=colors, marker=markers[i])
for j in range(3):
    ax.scatter([j+40]*len(cities), list(results_original_original.iloc[j+20])[1:], c=colors, marker=markers[j])
for k in range(2):
    ax.scatter([k+47.5]*len(cities), list(results_original_original.iloc[k+23])[1:], c=colors, marker=markers[k+1])


box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])
ax.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=3)
[plt.axvline(x, linestyle=':', c='gray') for x in [7, 17, 27, 37, 45, 51]]
plt.xticks([2, 12, 22, 32, 41, 48], models)
plt.title('Original features - Original dataset')
plt.xlim(-3,50.5)
plt.xlabel("Methods")
plt.ylabel("Weighted accuracy")
plt.show()
```
